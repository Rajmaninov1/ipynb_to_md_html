````python
file_to_open = "prueba regex/prueba_regex"
author = "Julian Tolosa"
date = "December 3, 2021"
````

````python
import json

with open(file_to_open+'.ipynb', encoding = "UTF-8") as file:
    data = json.load(file)
````

````python
programming_language = data["metadata"]["kernelspec"]
````

````python
cells = [ [x["cell_type"], x["source"], x["outputs"] if x["cell_type"] == "code" else ''] for x in data["cells"] ]
````

````python
yaml_config = f"""
---

<center>{file_to_open.split('/')[1]}</center>  
  <br />
author: {author}
  <br />
date: {date}
  <br />

---

"""

file_text = yaml_config
````

````python
for cell_type, source, output in cells:
    if cell_type == "markdown" and source:
        file_text += "".join(source) + "\n\n"
    elif cell_type == "code" and source:
        code_block = (
          f"""```\n{''.join(source)}\n```"""
        )
        if output != []:
            output_block = (
                f"""\n{''.join(output[0]["text"])}\n"""
            )
        else:
            output_block = ""
        file_text += code_block + output_block + "\n"
        
with open(file_to_open.split('/')[1]+".md", "w", encoding = "UTF-8") as file:
    file.write(file_text)
````
